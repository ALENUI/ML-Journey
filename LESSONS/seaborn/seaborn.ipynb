```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#visualizing the data
sns.set_theme()
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker', style='smoker', size='size', kind='scatter')
```

IRIS DATASET


```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
#scatter plot visualization
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', style='species', size='petal_length')
```

TITANIC DATASET


```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#count plot visualization
sns.countplot(data=titanic, x='class')
```

```python
sns.countplot(data=titanic, x='survived')
```

```python
#bar plot visualization
sns.barplot(data=titanic, x='sex', y='survived', hue='class')
```
