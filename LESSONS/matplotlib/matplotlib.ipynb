```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,10)
cos = np.cos(x)
sin = np.sin(x)
```

```python
plt.plot( sin, 'g--')
plt.plot( cos, 'r.')
plt.xlabel('angle')
plt.ylabel('cos/sin valye')
plt.title("Cosine and Sine Wave")
```

```python
#parabola
x= np.linspace(-10, 10, 20)
y=x**2
plt.plot(x,y, 'r+')
```
