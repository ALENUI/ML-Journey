```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,10)
cos = np.cos(x)
sin = np.sin(x)
```

```python
plt.plot( sin, 'g--')
plt.plot( cos, 'r.')
plt.xlabel('angle')
plt.ylabel('cos/sin valye')
plt.title("Cosine and Sine Wave")
```

```python
#parabola
x= np.linspace(-10, 10, 20)
y=x**2
plt.plot(x,y, 'r+')
```

##BAR PLOTS


```python
fig = plt.figure() #creates an emply plot
ax = fig.add_axes([0,0,1,1]) #0,0 sets the position of the plot and 1,1 sets the width and higfht of the plot

languages = ['English', 'French', 'Spanish' ]
people = [110, 76, 92]
ax.bar(languages, people)
plt.xlabel('Languages')
plt.ylabel("No of people")

```

Pie CHART


```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.pie(people, labels=languages, autopct='%1.1f%%')
```

Scatter Plot


```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.scatter( languages, people, color='g')
```

#D SCATTER PLOT


```python
fig=plt.figure()
ax=plt.axes(projection='3d')

z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)

ax.scatter(x,y,z, c=z, cmap="Blues")
plt.show()
```
